```python
%pylab inline
import matplotlib.pyplot as plt
import sys, os, glob
sys.path.append('/home/peter/code/pyTools/')

import numpy as np
from pyTools.system.videoExplorer import *
import os
from time import time

vE = videoExplorer()
fileList  = []
posList = []

t = time()
for root,  dirs,  files in os.walk('/run/media/peter/Elements/peter/data/tmp-20130506/'):
    files = files
    for f in files:
        if f.endswith('npy'):
            #fl = open(root + '/' + f, 'r')
            path = root + '/' + f
            fileList.append(path)
            posList.append(np.load(path))            

#sort both lists based on the file name
posList = [x for y, x in sorted(zip(fileList, posList))]
fileList = sorted(fileList)
print time() - t
```

```python
## smooth the positions
# first make long arrays for each vial and direction 

sequence =  [[] for i in range(8)]
for minute in posList:
    for pos in minute:
        for v in range(len(pos)):
            sequence[v * 2].append(pos[v][0])
            sequence[v * 2 + 1].append(pos[v][1])

```

```python
from scipy.signal import medfilt

filteredPos = []
for a in sequence:
    filteredPos.append(medfilt(a, 31))

```

```python
diff  =  []

for p in filteredPos:
    diff.append(np.convolve([1,-1], p))

dists = [[] for i in range(4)]
for i in range(4):
    dists[i] = np.sqrt(diff[i*2] ** 2 + diff[i*2 + 1] ** 2)
```

```python
# delete distance that are obvious errors
for d in dists:
    d[d > 25] = 0
```

```python
# accumulate distances for each minute
accDist = [[] for i in range(4)]
cnt = 0
for minute in posList:
    rng = slice(cnt, cnt+len(minute))
    for i in range(len(accDist)):
        accDist[i].append(np.sum(dists[i][rng]))
        
    cnt += len(minute)
    
accDist = np.asarray(accDist)
```

```python
def generateDistances(dists, vialNo):
    v = []
    
    for pos in dists:
        if pos != []:
            v.append(np.sum(pos[:,vialNo]))
        else:
            v.append(0.0)
    
    return v

# 1000px = 9.5cm = 0.095m
# 1px = 95mm

#fig, axs = plt.subplots(1,1, figsize=(20,10))


#%pylab inline 
#from matplotlib import rc
#rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
#rc('text', usetex=True)

t = time()
figure(figsize=(20,5))
color = ['r', 'g', 'y', 'b']
name = ['Ab +RU', 'Ab -RU', 'dilp', 'w+']
for i in range(3,-1,-1):
    v = accDist[i]
    
    smoothLen = 31.0
    m = np.convolve(v, np.ones(smoothLen)/smoothLen, 'same')
    
    #[val, filenames] = zip(*diff)
    
    plot(np.arange(m.shape[0]), m, color[i], label=name[i])
    
    ylim([0, 3500]) 
    xlim([0, m.shape[0]]) 

    
rng = range(0,m.shape[0], 60)
rng.append(m.shape[0]-1)

#axs.set_xticks(rng)
#axs.set_xticklabels([vE.fileName2DateTime(fileList[i], ending='npy') for i in rng], rotation=40, ha='right')


plt.xticks(rng, [vE.fileName2DateTime(fileList[i], ending='npy') for i in rng], rotation=45, ha='right')
plt.ylabel('Movement in mm / min')

spanPos = 2700
axvspan(spanPos, spanPos + 3, ymin=0, ymax=3500, color='pink', hold=None)

#plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%d' + r' $\frac{\mathrm{mm}}{\mathrm{min}}$'))
legend()

print time() - t
```

```python
    
def computeConfusionMatrix(predict, label):
    idx = np.unique(label)
    cmat = np.zeros((idx.shape[0], idx.shape[0]))
    
    for i in range(predict.shape[0]):
        cmat[label[i], predict[i]] += 1
        
    return cmat

from sklearn.cross_validation import StratifiedKFold
def crossValidate(data, labels, classifier,Nfolds=2):
    idx = np.unique(labels)
    cmat = np.zeros((idx.shape[0], idx.shape[0]), dtype=np.int32)
    
    #kf = KFold(len(labels), n_folds=Nfolds, indices=True)
    kf = StratifiedKFold(labels, n_folds=Nfolds, indices=True)
    
    cnt = 1
    
    predict = []
    testLbl = []
    testPos = []
    for train, test in kf:
        testPos.append(test)
        print("processing fold no {0}/{1}".format(cnt, Nfolds))
        
        trainSet = data[train]
        trainLbl = labels[train]
        
        testSet = data[test]
        testLbl.append(labels[test])
        
        wclf = classifier
        wclf.fit(trainSet, trainLbl)
        
        predict.append(wclf.predict(testSet))
        
        cmat += computeConfusionMatrix(predict[-1], testLbl[-1])
        cnt += 1
    
    return cmat, predict, testLbl, testPos

def getMissClassified(predict, labels, pathList, cl):
    out = []
    
    for i in range(len(pathList)):
        if predict[i] != labels[i] and labels[i] == cl:
            out.append(pathList[i])
            
    return out
```
